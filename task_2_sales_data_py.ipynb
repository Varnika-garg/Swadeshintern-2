Data Loading and cleaning


*   load data: use panda to load the sales data



```python
import pandas as pd
sales_data=pd.read_csv(r"/content/Sales_data_.csv")
print(sales_data)
```



*   CLEAN Data:check for missing values,duplicate rows or inconsistances and handle them appropriately.




```python
sales_data.dropna(inplace=True)
sales_data.drop_duplicates(inplace=True)
```

2-Data analysing

```python
#basic matrics
total_sales=sales_data['sales'].sum()
average_sales=sales_data['sales'].mean()
sales_per_product=sales_data.groupby('product')['sales'].sum()
print(total_sales)
print(average_sales)
print(sales_per_product)
```

```python
import matplotlib.pyplot as plt
sales_data['date']=pd.to_datetime(sales_data['date'])
sales_data.set_index('date',inplace=True)
monthly_sales=sales_data['sales'].resample('M').sum()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index,monthly_sales,marker='o')
plt.title('Monthly Sales trends')
plt.xlabel('date')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

3-Identify trends and top performing products

```python
#trends:Analyze how sales changes over different time period
quarterly_sales=sales_data['sales'].resample('Q').sum()
print(quarterly_sales)
```

```python
#top performing products
top_products=sales_per_product.sort_values(ascending=False)
print(top_products)
```
